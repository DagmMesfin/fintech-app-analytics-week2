# Aspect-Based Sentiment Analysis (ABSA)
Extend the analysis from overall sentiment to aspect-based sentiment tied to features (e.g., login, speed, reliability).

This section:
- Defines aspect keyword groups
- Extracts aspects from each review
- Classifies sentiment per aspect using the existing Transformers classifier from `sentiment_analysis`
- Produces a long-form table for aggregation and visualization

```python
# Define aspect keywords (customize as needed)
ASPECT_KEYWORDS = {
    "login": ["login", "log in", "signin", "sign in", "password", "otp"],
    "speed": ["slow", "lag", "fast", "quick", "speed", "responsive", "delay"],
    "reliability": ["reliable", "unreliable", "downtime", "outage", "server down", "stable", "unstable"],
    "performance": ["performance", "crash", "freeze", "hang", "bug", "buggy", "glitch"],
    "ui": ["ui", "ux", "interface", "design", "layout", "user friendly", "navigation"],
    "payment": ["payment", "pay", "bill", "merchant", "qr", "pos"],
    "transfer": ["transfer", "send", "receive", "transaction", "funds", "remit"],
    "security": ["secure", "security", "safe", "fraud", "scam", "pin", "biometric"],
    "notification": ["notification", "alert", "reminder"],
    "update": ["update", "version", "upgrade"],
    "install": ["install", "installation", "download"],
    "account": ["account", "balance", "statement", "history", "limit"],
    "customer_support": ["support", "help", "service", "response", "call", "email", "feedback"],
}
print("Defined", len(ASPECT_KEYWORDS), "aspect groups")
```

```python
import re
from statistics import mean

# Lightweight sentence splitter

def sentence_split(text: str):
    return [s.strip() for s in re.split(r"[.!?]+", str(text)) if s.strip()]


def normalize(text: str) -> str:
    return str(text).lower()


def extract_aspects(text: str, aspect_keywords: dict) -> list:
    t = normalize(text)
    found = []
    for aspect, kws in aspect_keywords.items():
        for kw in kws:
            if kw in t:
                found.append(aspect)
                break
    return sorted(set(found))


def score_text_sentiment(text: str, classifier):
    try:
        res = classifier(str(text))[0]
        return {"label": res.get("label", "NEUTRAL"), "score": float(res.get("score", 0.0))}
    except Exception:
        return {"label": "NEUTRAL", "score": 0.0}


def aspect_sentiment_for_review(
    text: str,
    aspects: list,
    aspect_keywords: dict,
    classifier,
    neutral_threshold: float = 0.0,
):
    results = {}
    sentences = sentence_split(text)
    for aspect in aspects:
        kws = aspect_keywords.get(aspect, [])
        target_sentences = [s for s in sentences if any(kw in normalize(s) for kw in kws)]
        if not target_sentences:
            target_sentences = [text]  # fallback to whole review

        scored = [score_text_sentiment(s, classifier) for s in target_sentences]
        pos_scores = [s["score"] for s in scored if s["label"].upper().startswith("POS")]
        neg_scores = [s["score"] for s in scored if s["label"].upper().startswith("NEG")]

        if not pos_scores and not neg_scores:
            results[aspect] = {"label": "NEUTRAL", "score": 0.0}
            continue

        avg_pos = mean(pos_scores) if pos_scores else 0.0
        avg_neg = mean(neg_scores) if neg_scores else 0.0
        if max(avg_pos, avg_neg) < neutral_threshold:
            results[aspect] = {"label": "NEUTRAL", "score": max(avg_pos, avg_neg)}
        else:
            label = "POSITIVE" if avg_pos >= avg_neg else "NEGATIVE"
            results[aspect] = {"label": label, "score": max(avg_pos, avg_neg)}
    return results
```

```python
# Select a source DataFrame: prefer thematic_analysis.df, else sentiment_analysis.df
source_df = None
if "thematic_analysis" in globals() and hasattr(thematic_analysis, "df"):
    source_df = thematic_analysis.df.copy()
elif "sentiment_analysis" in globals() and hasattr(sentiment_analysis, "df"):
    source_df = sentiment_analysis.df.copy()

if source_df is None:
    raise RuntimeError("No source DataFrame found. Initialize sentiment_analysis or thematic_analysis first.")

# Ensure expected columns
if "review" not in source_df.columns:
    raise RuntimeError("Source DataFrame must contain a 'review' column.")

# Extract aspects and compute aspect-level sentiments
source_df["aspects"] = source_df["review"].apply(lambda x: extract_aspects(x, ASPECT_KEYWORDS))
source_df["aspect_sentiments"] = source_df.apply(
    lambda r: aspect_sentiment_for_review(
        r["review"], r["aspects"], ASPECT_KEYWORDS, sentiment_analysis.classifier
    ) if r["aspects"] else {},
    axis=1,
)

# Create a long-form table for aggregation and visualization
rows = []
for idx, r in source_df.iterrows():
    bank = r["bank"] if "bank" in source_df.columns else None
    rating = r["rating"] if "rating" in source_df.columns else None
    for aspect in r["aspects"]:
        s = r["aspect_sentiments"].get(aspect, {"label": "NEUTRAL", "score": 0.0})
        rows.append({
            "review_id": idx,
            "bank": bank,
            "aspect": aspect,
            "aspect_label": s["label"],
            "aspect_score": s["score"],
            "rating": rating,
        })

import pandas as pd
aspect_sentiments_long = pd.DataFrame(rows)
print("ABSA complete. Sample:")
display(aspect_sentiments_long.head(10))
```

```python
# Simple aggregations
if not aspect_sentiments_long.empty:
    agg_counts = aspect_sentiments_long.groupby(["aspect", "aspect_label"]).size().reset_index(name="count")
    agg_scores = aspect_sentiments_long.groupby(["aspect"])["aspect_score"].mean().reset_index(name="avg_aspect_score")
    print("Counts by aspect/label:")
    display(agg_counts.sort_values(["aspect", "count"], ascending=[True, False]).head(20))
    print("Average confidence score by aspect:")
    display(agg_scores.sort_values("avg_aspect_score", ascending=False))
else:
    print("No aspects detected with current keywords. Consider expanding ASPECT_KEYWORDS.")
```
